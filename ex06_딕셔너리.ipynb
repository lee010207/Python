## 딕셔너리 dictionary


```python
# 딕셔너리 생성
# 딕셔너리명 = {Key:value,Key:value , , }
# Key값은 변하지 않을 값, 동일한 Key는 여러개 올 수 없다(중복 안돼)

a = {}
print(type(a))
print(a)

b = {'name' :'Joy' , 'pn':'010-1234-5678'}
c = {1:'홍길동', 2:'김철수'}

print(b)
print(c)
```

```python
# 딕셔너리 값 추가
# 딕셔너리명[추가할 key값] = 추가할 value값
dic1 ={'name':'홍길동', 'age':20 , 'phone':'010-2394-0837'}
dic1['birth'] = '03/15'

# 딕셔녀리 값 변경
# 원래 딕셔너리에 있던 키값을 이용하면 값이 변경됨
# 딕셔너리명[변경할 키값] = 변경할 벨류값

dic1['name'] = '고길동'
print(dic1)
```

```python
# 딕셔너리 값 가져오기 (1)
print(dic1['name'])
print(dic1['birth'])

# 리스트 안에 여러개의 딕셔너리를 담아서 인덱싱 활용해서 하나의 주소록을 관리
# 사용자한테 입력받아서 append 시켜주면 입력받아서 바로 주소록에 저장 가능
data = [{'name':'홍길동', 'age':20 , 'phone':'010-2394-0837'}, {'name':'김조이', 'age':53 , 'phone':'010-5474-2386'}]
data.append({})
print(data[0]['name'])
```

```python
# 딕셔너리 값 가져오기 (2)
# 딕셔너리명.get(key)

dic1.get('age')

# 없는 키값을 넣었을 때 -> .get() 은 오류 안나
s = dic1.get('addr')
print(s)
# dic1['addr']  --> 오류나
```

```python
# 딕셔너리 key만 가져오기
print(dic1.keys())

#리스트로 형변환 : 리스트에 담아주기
klist = list(dic1.keys())
print(klist)
```

```python
# 딕셔너리 value만 가져오기
print(dic1.values())

vlist = list(dic1.values())
print(vlist)
```

```python
# 딕셔너리 key, value 동시에 가져오기
print(dic1.items())

kvlist = list(dic1.items())
print(kvlist)
```

```python
# 값 삭제하기
# del 딕셔너리명[key]

del dic1['age']
print(dic1)
```

```python
# 전체 삭제
# 딕셔너리명.clear()

dic1.clear()
print(dic1)
```

```python
# 딕셔너리를 멤버연산자와 같이 사용하기
# 딕셔너리 안에 특정 key값이 포함되어있는지 확인

20 in dic1.values()
```
