## 문자열 포매팅(Formatting)



```python
# 문자열 안에 특정 값을 바꿔야 할 경우
month = 7
day = 3

# 문자열과 숫자형 데이터는 더하기 연산이 불가
s= '오늘은', month,'월',day,'일 입니다.'
print(s)
```

```python
# - %기호 포매팅

s = '오늘은 %d월 %d일 입니다.'%(month, day)
print(s)
```

```python
a = 3.12
b = 27.3192
c = '%f 더하기 %f는 %f입니다'%(a, b, a+b)
print(c)
```

```python
# format 함수를 이용한 포매팅

s = '오늘은 {}월 {}일 입니다.'.format(month, day)
print(s)
```

```python
# f 문자열을 이용한 포매팅

s = f'오늘은 {month}월 {day}일 입니다.'
print(s)
```

```python
# 변수 예제
x = 100
y = 200
sum = x + y
print('{}와 {}의 합은 {}입니다.'.format(x, y, sum))
print ('%d와 %d의 합은 %d입니다.'%(x, y, sum))
print(f'{x}와 {y}의 합은 {sum}입니다.')
```

```python
# 문자 개수 세기
# 문자열.count(데이터)
t = 'hello hi python world'
print(t.count('l'))
```

```python
# 위치 알려주기

# 문자열.find(데이터)
# 제일 먼저 나오는 위치를 찾아줌
# 찾는 문자가 존재하지 않을 경우 -1 반환
print(t.find('l'))

# 문자열.index(데이터)
print(t.index('e'))

# 찾는 데이터 값이 없을 때
# --> index : 오류 발생
# --> find : -1 반환
```

```python
# 문자열 삽입
# '문자'.join(문자열)

print('~'.join(t))
```

```python
# 문자열 나누기
# 문자열.split(구분문자)
t.split()
# space tab enter 기준으로 나눠줌 : () 만 쓸 경우
```
