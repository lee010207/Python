##  (Formatting)



```python
#       
month = 7
day = 3

#      
s= '', month,'',day,' .'
print(s)
```

```python
# - % 

s = ' %d %d .'%(month, day)
print(s)
```

```python
a = 3.12
b = 27.3192
c = '%f  %f %f'%(a, b, a+b)
print(c)
```

```python
# format   

s = ' {} {} .'.format(month, day)
print(s)
```

```python
# f   

s = f' {month} {day} .'
print(s)
```

```python
#  
x = 100
y = 200
sum = x + y
print('{} {}  {}.'.format(x, y, sum))
print ('%d %d  %d.'%(x, y, sum))
print(f'{x} {y}  {sum}.')
```

```python
#   
# .count()
t = 'hello hi python world'
print(t.count('l'))
```

```python
#  

# .find()
#     
#      -1 
print(t.find('l'))

# .index()
print(t.index('e'))

#     
# --> index :  
# --> find : -1 
```

```python
#  
# ''.join()

print('~'.join(t))
```

```python
#  
# .split()
t.split()
# space tab enter   : ()   
```
