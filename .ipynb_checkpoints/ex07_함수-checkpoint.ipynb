```python
# 함수 정의 후 실행됐는지 확인
# 함수 정의 / 함수 실행 셀 따로 분리 권장

def number_sum(num1, num2):
    return num1 + num2
```

- 예제


```python
def cal(num1, num2, op):
    if op == "+":
        result = num1+num2
        
    elif op == '-':
        result = num1 - num2
    return result
```

```python
num1 = int(input("첫번째 정수 입력 >> "))
num2 = int(input("두번째 정수 입력 >> "))
op = input("연산자 입력(+,-) >> ")
result = cal(num1, num2, op)

print("결과 : {}".format(result))
```

- 답안 코드

```python
def cal(num1, num2, op):
    if op == "+":
        return num1+num2
        
    elif op == '-':
        return num1 - num2

# return 값이 될 수도 있고(+,-)
# return 값이 안될 수고 있음(+,- 외 다른 값 : * , /, % ...) --> 오류 안나
```

```python
def cal(num1, num2, op):
    """ 덧셈과 뺄셈을 계산하는 함수 """
    # docstring : 함수에 대한 설명을 달아줄 때 """ """
    
    if op == "+":
        return num1+num2
        
    elif op == '-':
        return num1 - num2
    else :
        return '연산기호 입력 에러'

# 정수형 타입, 문자열 타입 return : 여러 데이터 타입을 리턴 가능
```

```python
num1 = int(input("첫번째 정수 입력 >> "))
num2 = int(input("두번째 정수 입력 >> "))
op = input("연산자 입력(+,-) >> ")

print("결과 : {}".format(cal(num1,num2,op)))
```

```python
# 매개변수 기본값 설정하기
# 함수 정의할 때 특정 매개변수의 값(기본값)을 지정해주기

def cal(num1, num2, op = '+'):
    if op == "+":
        return num1+num2
        
    elif op == '-':
        return num1 - num2
```

```python
cal(15,2)
# op 가 기본값으로 + 니까 안써줘도(생략해도) 오류 안나고 더하기값으로 반환
```

```python
# 다른 값을 반환받고 싶을 때는 매개변수 자리에 값을 넣어주면 됨
cal(13,2,'-')
```

```python
# num2, op 에 각각 기본값 지정

def cal(num1, num2 = 1, op = '+'):
    if op == "+":
        return num1+num2
        
    elif op == '-':
        return num1 - num2
```

```python
cal(6, op = '-')
# num2는 기본값으로 하고 op만 '-'로 바꿔주고 싶으면 
# 매개변수명을 명시해서 대입연산자 이용해서 값 지정해주면 됨
```

```python
print(cal(11))
# num1 = 11, num2 = 1(기본값), op = '+'(기본값)

# 매개변수 타입이 정해져 있지 않음 --> 데이터 타입에 상관없이 호출 가능
print(cal('A', 'B'))
```

### 가변 매개변수

#### 메서드 매개변수를 가변적으로 받을 수 있음


```python
# 매개변수에 전달된 모든 값을 반환하는 함수 만들기

def add(*args):
    print(args)
```

```python
add(1,2,3)
add('a','b','c',1,2,3)
```

```python
# 매개변수에 전달된 모든 값의 합을 반환하는 함수 만들기

def add(*args):
    total = 0
    for n in args:
        total += n
    return total
```

```python
print(add(1,2,3))
print(add(4,5,6,7))
```

```python
# 데이터 타입 : 튜플
def add(*a):
    print(type(a))
```

```python
add(1)
```

```python
def add(num, *args):
    total = 0
    for n in args:
        total += n
    return total
# 가변 매개변수랑 일반 매개변수를 섞어 사용할 때
# --> 일반 매개변수 먼저!!
```

```python
print(add(1,2,3))
print(add(4,5,6,7))
```

```python
# 가변 매개변수가 먼저 오면 에러남

def add(*args, num):
    total = 0
    for n in args:
        total += n
    return total
```

```python
print(add(1,2,3))
```

```python
# return 키워드는 한번만 사용가능!

def add_sub(n1, n2):
    return "더하기 빼기 결과",n1+n2, n1-n2
```

```python
print(add_sub(3,9))

# 튜플 안에 담겨서 값이 반환
# 데이터 타입 섞여도 가능
```

```python
str1, ar, sr = add_sub(3,9)

print(str1)
print(ar)
print(sr)
```

```python
# 변수 담을 때 이렇게도 가능했어
a, b, c = 1, 2, 3
print(a)
print(b)
print(c)
```

```python
a, b, c = [1, 2, 3]
print(a)
print(b)
print(c)

print(type(a))

s = [1,2,3]
print(type(s))
```

```python
# 튜플 만들때는 소괄호 생략 가능!!
re = 7,3
print(re)
print(type(re))
```
