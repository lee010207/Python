## 반복문


- while 문


```python
while True:
    num1 = int(input("첫 번째 정수 입력 : "))
    num2 = int(input("두 번째 정수 입력 : "))
    if num1 == 0 and num2 == 0 :
        print("프로그램 종료")
        break
    else:
        print("두 정수의 합 : {}".format(num1+num2))
```

```python
import random

randomNumber = random.randint(1,50)
print(randomNumber)


while True:
    inputNum = int(input("숫자를 입력하세요 >> "))
    if randomNumber > inputNum :
        print("{}보다 큰 수 입니다.".format(inputNum))
    elif randomNumber < inputNum :
        print("{}보다 작은 수 입니다.".format(inputNum))
    elif randomNumber == inputNum :
        print("정답을 맞추셨습니다.")
        break
```

- for문


```python
lst_food = ["삼겹살", "파스타","짜장면","계란말이"]
for food in lst_food :
    print(food)
```

```python
score_list = [90,45,70,60,55,90]
cnt = 1
for i in score_list:                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  
    if i >= 60 :
        print("{}번 학생은 합격입니다.".format(cnt))
    else :
        print("{}번 학생은 불합격입니다.".format(cnt))
    cnt += 1
```

```python
list3 = [-5,-2, -13 ,-57]

minNum = list3[0]
for i in range(len(list3)) :
    if minNum > list3[i]:
        minNum = list3[i]
print(minNum)
```

```python
import pickle
```

```python
# pickle 모듈을 이용해 데이터 로드
# 외부 파이썬 파일 을 불러가 쓰는 방식
# with open('파일 경로/파일명/확장자' , 'rb'--> 읽기 전용) as 별칭
# tab 이용해서 자동완성으로 써주면 오타 줄일 수 있음

with open('data/movie_review.pkl' , 'rb') as f :
    movie_review = pickle.load(f)
```

```python
# 1 : 긍정, 0 : 부정

# 전체 리뷰 개수 확인
print(len(movie_review))
# 3번째 리뷰 데이터 가져오기
print(movie_review[2])
# 3번째 리뷰에서 텍스트(리스트 첫번째 값)만 가져오기
print(movie_review[2][0])
# 3번째 리뷰에서 텍스트(리스트 두번째 값)만 가져오기
print(movie_review[2][1])
```

```python
sum = 0
for i in range(len(movie_review)):
    sum += len(movie_review[i][0])
print("리뷰의 평균 길이는 '{}'입니다.".format(sum/len(movie_review)))
```

```python
pos =0;
neg = 0;

for i in range(len(movie_review)) :
    if movie_review[i][1] == 1 : 
        pos += 1
    else :
        neg += 1
print("긍정 '{}'개".format(pos))
print("부정 '{}'개".format(neg))
```

```python
review = 0
for txt, pn in movie_review :
    # pn값이 0 아니면 1이고, 0은 더해도 그대로고 1만 카운팅되는 거랑 똑같
    review += pn
print("긍정 '{}'개".format(review))
print("부정 '{}'개".format(len(movie_review)-review))
```

### 리뷰 최대 값/리뷰 내용 출력하기

```python
maxReview = movie_review[0][0]

for txt, pn in movie_review :
    if len(maxReview) < len(txt):
        maxReview = txt

print("리뷰 최대길이는 '{}'입니다".format(len(maxReview)))
print(maxReview)

## for i in range(len(리스트, 튜플, 문자열)) --> 인덱스 값 활용
## for 변수 in 리스트,튜플,문자열 --> 안에 들어있는 데이터, 값 활용
```

```python
search = input("검색할 데이터 입력 >> ")
cnt = 0
for txt , pn in movie_review :
    if search in txt:
        cnt += 1
print(search + "가 포함된 리뷰는 {}개 입니다.".format(cnt))
```
