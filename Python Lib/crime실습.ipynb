```python
import pandas as pd
```

```python
c15 = pd.read_csv('2015.csv' , encoding='euc-kr' ,index_col='')
c16 = pd.read_csv('2016.csv' , encoding='euc-kr',index_col='')
c17 = pd.read_csv('2017.csv' , encoding='euc-kr',index_col='')
c15
```

```python
cm15 = c15[c15['']=='']
a15 = cm15.loc[:,'':''].sum(axis=1)
a15
```

```python
cm16 = c16[c16['']=='']
a16 = cm16.loc[:,'':''].sum(axis=1)
a16
```

```python
c17 = c17.drop('')
cm17 = c17[c17['']=='']
a17 = cm17.loc[:,'':''].sum(axis=1)
a17
```

```python
h1516=(a16-a15)/a15*100
h1516
```

```python
h1617=(a17-a16)/a16*100
h1617
```

```python
result = pd.concat([a15,h1516,a16,h1617,a17],  axis=1 , keys=['2015', '2015-2016','2016','2016-2016','2017'])
```
