```python
import matplotlib.pyplot as plt
```

```python
y = [2,4,6]

# plot : 선 그래프(Line plot)
plt.plot(y)
plt.show()
```

```python
y = [2,4,6]
x = [1,2,3]
# x,y 쌍이 맞아야 에러가 안남 : 각 x좌표, y좌표 쌍을 찍어서 그래프로 나타내주기 때문
# x,y 개수가 일치해야 함!!
plt.plot(x,y)
plt.show()
```

```python
# 그래프 스타일 적용하기
# line style 매개변수 활용 : ls
# -(default), -- , : , -.
plt.plot(x,y , linestyle = '-.')
# plt.plot(x,y, ls=':')
plt.show()
```

```python
# Marker : 특정 위치(좌표) 표시
plt.plot(x,y , marker = 'D')
plt.show()
```

```python
x = [2,4,6,8]
y = [10,13,16,20]
y2 = [10,20,15,12]

plt.plot(x,y, ls=':', marker='o')
plt.plot(x,y2 , marker='^')
plt.show()
```

```python
# line width 선 두께 : lw
plt.plot(x, y, linewidth=3)
# plt.plot(x, y, lw=3)
plt.show()
```

```python
# line color 변경 : color
plt.plot(x,y , color='#8BE8E5', marker='p')
plt.show()
```

```python
# marker color 마커 색상 : mfc / makerfacecolor 
plt.plot(x,y , color='#8BE8E5', marker='p', markerfacecolor='#102C57')
plt.show()
```

```python
# 마커 선(테두리) 색상 : markeredgecolor / mec
plt.plot(x,y , color='#8BE8E5', marker='p', markerfacecolor='#F7C8E0', mec='#102C57')
plt.show()
```

### 실습


```python
import numpy as np
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', lw='3',color='#B6EAFA', marker='o',
         mfc='#FFEBEB',mec='#FCFFB2',ms='10', mew='3')

# 그래프 범위 지정 : 축(가로 x축(xlim), 세로 y축(ylim)) 값 조절
plt.xlim(-1,8)
plt.ylim(-1,11)
plt.show()
```

```python

# 틱의 위치를 보기 위해, 격자무늬 표현
# tick : 그래프 축의 위치
plt.plot(x,y, ls='--', lw='3',color='#B6EAFA', marker='o',
         mfc='#FFEBEB',mec='#FCFFB2',ms='10', mew='3')
plt.xlim(-1,8)
plt.ylim(-1,10)
plt.xticks([1,3,5,7])
plt.yticks([2,4,6,8])
# x축 틱 설정
plt.grid()
plt.show()
```

```python
# 선그래프 plot 여러개 그리기
x = [3,7,11,6,13,10]
y = [2,5,10,15,3,8]
z = [1,13,7,4,12,9]

# label : 그래프에 이름 달아주기
plt.plot(x , label='x')
plt.plot(y , label='y')
plt.plot(z , label='z')

# 그래프 범례 출력
plt.legend()
plt.show()
```

```python
plt.plot(x , label='엑스' ,color='#002B5B')
plt.plot(y , label='와이' ,color='#F0EB8D')
plt.plot(z , label='제트' ,color='#97DEFF')

# 한글로 이름을 달아주려면
# 한글을 지원해주는 폰트로 변경해줘야 함!
from matplotlib import rc
rc('font', family='Malgun Gothic')

plt.legend()
# 축에 이름 표시
plt.xlabel('x축')
plt.ylabel('y축', rotation=0)

#그래프릐 이름
plt.title('x,y,z 그래프')

plt.show()
```
