```python
import pandas as pd
```

```python
c15 = pd.read_csv('2015.csv' , encoding='euc-kr' ,index_col='관서명')
c16 = pd.read_csv('2016.csv' , encoding='euc-kr',index_col='관서명')
c17 = pd.read_csv('2017.csv' , encoding='euc-kr',index_col='관서명')
c15
```

```python
cm15 = c15[c15['구분']=='발생건수']
a15 = cm15.loc[:,'살인':'폭력'].sum(axis=1)
a15
```

```python
cm16 = c16[c16['구분']=='발생건수']
a16 = cm16.loc[:,'살인':'폭력'].sum(axis=1)
a16
```

```python
c17 = c17.drop('광주지방경찰청')
cm17 = c17[c17['구분']=='발생건수']
a17 = cm17.loc[:,'살인':'폭력'].sum(axis=1)
a17
```

```python
h1516=(a16-a15)/a15*100
h1516
```

```python
h1617=(a17-a16)/a16*100
h1617
```

```python
result = pd.concat([a15,h1516,a16,h1617,a17],  axis=1 , keys=['2015총계', '2015-2016증감율','2016총계','2016-2016증감율','2017총계'])
```
