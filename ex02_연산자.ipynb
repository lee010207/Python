```python
# java Scanner 처럼 키보드로 데이터 입력받기
# input()

num = input("정수입력 >> ")
print(num)
print(type(num))
# num이 문자형이라 더하기 연산이 안되니까
# 강제 형변환 해줌
print(int(num)+1111)
```

```python
num1 = int(input("입력 : "))
num2 = int(input("입력 : "))
sum = num1 + num2
sub = num1 - num2
x = num1*num2
na = num1/num2
print('더하기 결과 : ', sum)
print('빼기 결과 : ', sub)
print('곱하기 결과 : ', x)
print('나누기 결과 : ', na)
```

```python
time = int(input("시간 입력 >> "))
hour = time//3600
minute = time%3600//60
second = time%3600%60
print("{}시간 {}분 {}초".format(hour, minute, second))
```

- 삼항연산자



```python
score = 80
result = '합격' if score>=60 else '불합격'
print(result)
```

```python
num = int(input("정수 입력 >> "))
result = "짝수" if num % 2 == 0 else "홀수"
print("{}은(는) ".format(num) + result + "입니다.")
```
