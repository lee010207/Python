### 조건문


```python
if True :
    print('true면 실행문 1')
    print('실행문 2')
    # if문 안에 if문 가능
    if False :
        print('실행문 3')
    print('실행문 4')
print('false면 조건문 종료')
# 들여쓰기로 if문 구분!! --> 실행문들이 들여쓰기 라인이 똑같아야 에러가 안남
```

```python
score = int(input("점수를 입력하세요 >> "))
if 90 <= score <= 100 :
    print("{}점은 A학점 입니다".format(score))
elif score >= 80 :
    print("{}점은 B학점 입니다".format(score))
elif score >= 70 :
    print("{}점은 C학점 입니다".format(score))
elif score >= 60 :
    print("{}점은 D학점 입니다".format(score))
else :
    print("{}점은 F학점 입니다".format(score))
```

### 실습


```python
print('------ SM Coffee ------'+'\n')
money = int(input("금액을 입력하세요 >> "))
print('커피를 고르세요')
select = int(input('[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))
if select == 1:
    if(money >= 600):
        print("거스름돈 : ", (money - 600),"원")
    else :
        print('돈이 부족합니다.')
        print("거스름돈 : ", money,"원")
elif select == 2:
    if(money >= 1000):
        print("거스름돈 : ", (money - 1000),"원")
    else :
        print('돈이 부족합니다.')
        print("거스름돈 : ", money)
elif select == 3:
    if(money >= 1300):
        print("거스름돈 : ", (money - 1300),"원")
    else :
        print('돈이 부족합니다.')
        print("거스름돈 : ", money,"원")
else :
    print("잘못 입력하셨습니다.")
```

- 실습 코드 답안

```python
print('-----[SM Coffee]-----')
inputMoney = int(input('금액을 입력하세요 >> '))
print('커피를 고르세요.')
menu = int(input('[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))

if menu == 1 :
    price = 600
elif menu == 2 :     
    price = 1000
elif menu == 3 : 
    price = 1300

if inputMoney < price :
    print('돈이 부족해요..T^T')
    price = 0
    
print(f'거스름돈 : {inputMoney - price}원')
```
