```python
def cal(num1, num2, op):
    """     """
    # docstring :      """ """
    
    if op == "+":
        return num1+num2
        
    elif op == '-':
        return num1 - num2
    else :
        return '  '

#  ,   return :     
```

```python
num1 = int(input("   >> "))
num2 = int(input("   >> "))
op = input(" (+,-) >> ")

print(" : {}".format(cal(num1,num2,op)))
```

```python
#   
#      () 

def cal(num1, num2, op = '+'):
    if op == "+":
        return num1+num2
        
    elif op == '-':
        return num1 - num2
```

```python
cal(15,2)
# op   +  ()    
```

```python
#          
cal(13,2,'-')
```

```python
# num2, op    

def cal(num1, num2 = 1, op = '+'):
    if op == "+":
        return num1+num2
        
    elif op == '-':
        return num1 - num2
```

```python
cal(6, op = '-')
# num2   op '-'   
#       
```

```python
print(cal(11))
# num1 = 11, num2 = 1(), op = '+'()

#      -->     
print(cal('A', 'B'))
```
